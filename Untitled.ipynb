```python
bike_data_hour =  pd.read_csv('Bike-sharing-dataset/hour.csv')
bike_data_day =  pd.read_csv('Bike-sharing-dataset/day.csv')
```

```python
bike_data_day.info()
```

```python
# Mengkonversikan tipe object pada dteday menjadi datetime
bike_data_day['dteday'] = pd.to_datetime(bike_data_day['dteday'])
```

```python
bike_data_day.info()
```

```python
# Membuat feature baru
bike_data_day['day_of_week'] = bike_data_day['dteday'].dt.day_name()
```

```python
# Plot tren harian
plt.figure(figsize=(12, 6))
daily_trend = bike_data_day.groupby('dteday')['cnt'].sum()
daily_trend.plot()
plt.title('Tren Harian Peminjaman Sepeda')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Peminjaman')
plt.show()
```

```python
# Plot perbedaan antar musim
plt.figure(figsize=(12, 6))
seasonal_diff = bike_data_day.groupby('season')['cnt'].sum()
seasonal_diff.index = ['Spring', 'Summer', 'Fall', 'Winter']
seasonal_diff.plot(kind='bar', color=['green', 'blue', 'orange', 'purple'])
plt.title('Perbedaan Peminjaman Sepeda antar Musim')
plt.xlabel('Musim')
plt.ylabel('Jumlah Peminjaman')
plt.show()
```

```python
# Plot perbandingan antara tahun 2011 dan 2012
plt.figure(figsize=(12, 6))
yearly_comparison = bike_data_day.groupby('yr')['cnt'].sum()
yearly_comparison.index = ['2011', '2012']
yearly_comparison.plot(kind='bar', color=['lightcoral', 'skyblue'])
plt.title('Perbandingan Peminjaman Sepeda antara Tahun 2011 dan 2012')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Peminjaman')
plt.show()
```

```python
weather_columns = ['temp', 'atemp', 'hum', 'windspeed', 'cnt']
weather_data = bike_data_day[weather_columns]
```

```python
# Menghitung korelasi antara variabel cuaca dan jumlah peminjaman
correlation_matrix = weather_data.corr()

# Plot matriks korelasi menggunakan heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriks Korelasi Antara Variabel Cuaca dan Jumlah Peminjaman Sepeda')
plt.show()

# Analisis perbedaan kondisi cuaca terhadap jumlah peminjaman
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.scatterplot(x='temp', y='cnt', data=bike_data_day, alpha=0.5)
plt.title('Pengaruh Temperatur terhadap Jumlah Peminjaman')

plt.subplot(2, 2, 2)
sns.scatterplot(x='hum', y='cnt', data=bike_data_day, alpha=0.5)
plt.title('Pengaruh Kelembaban terhadap Jumlah Peminjaman')

plt.subplot(2, 2, 3)
sns.scatterplot(x='windspeed', y='cnt', data=bike_data_day, alpha=0.5)
plt.title('Pengaruh Kecepatan Angin terhadap Jumlah Peminjaman')

plt.subplot(2, 2, 4)
sns.boxplot(x='weathersit', y='cnt', data=bike_data_day)
plt.title('Perbandingan Kondisi Cuaca terhadap Jumlah Peminjaman')

plt.tight_layout()
plt.show()
```

```python
# Plot pola peminjaman pada berbagai jam dalam sehari
plt.figure(figsize=(12, 6))
hourly_trend = bike_data_hour.groupby('hr')['cnt'].mean()
hourly_trend.plot(kind='bar', color='skyblue')
plt.title('Pola Peminjaman Sepeda pada Berbagai Jam dalam Sehari')
plt.xlabel('Jam dalam Sehari')
plt.ylabel('Rata-rata Jumlah Peminjaman')
plt.show()

# Puncak dan lembah peminjaman sepeda selama 24 jam
plt.figure(figsize=(12, 6))
sns.lineplot(x='hr', y='cnt', data=bike_data_hour, errorbar=None)
plt.title('Puncak dan Lembah Peminjaman Sepeda Selama 24 Jam')
plt.xlabel('Jam dalam Sehari')
plt.ylabel('Jumlah Peminjaman')
plt.show()
```
